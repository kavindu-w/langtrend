```python
from pathlib import Path
import json

html_cache_dir = Path("../data/processed/html_cache")
lang_data_path = Path("../data/processed/language_data.json")

# Build language -> class_id lookup
lang_to_class = {}
if lang_data_path.exists():
    with open(lang_data_path, "r", encoding="utf-8") as f:
        lang_data = json.load(f)
    for class_id, langs in lang_data["lang_classes"].items():
        for lang in langs:
            lang_to_class[lang] = int(class_id)
    print(f"Loaded {len(lang_to_class)} language->class mappings\n")
else:
    print(f"Warning: language_data.json not found at {lang_data_path}\n")

files_with_languages = []

if html_cache_dir.exists():
    json_files = sorted(html_cache_dir.glob("*.json"))
    print(f"Found {len(json_files)} JSON files in html_cache/\n")

    for json_file in json_files:
        paper_id = json_file.stem
        with open(json_file, "r", encoding="utf-8") as f:
            data = json.load(f)

        paper_sections = []
        for section_name, section_data in data.items():
            detected = section_data.get("detected", [])
            filtered = sorted(set(lang for lang in detected if lang_to_class.get(lang, 5) < 5))
            if filtered:
                paper_sections.append({"section": section_name, "languages": filtered})
                files_with_languages.append({
                    "paper_id": paper_id,
                    "file": str(json_file),
                    "section": section_name,
                    "languages": filtered,
                })

        if paper_sections:
            print(f"Paper ID:  {paper_id}")
            print(f"File:      {json_file}")
            for s in paper_sections:
                print(f"  {s['section']}: {len(s['languages'])} unique — {s['languages']}")
            print()

    print(f"\n=== SUMMARY ===")
    print(f"Total sections with detected languages (class 0–4): {len(files_with_languages)}")
else:
    print(f"Directory not found: {html_cache_dir}")
```
